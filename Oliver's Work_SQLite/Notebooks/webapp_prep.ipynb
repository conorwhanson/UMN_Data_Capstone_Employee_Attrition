```python
import pandas as pd
import numpy as np
import pickle
from sqlalchemy import create_engine, inspect
```

```python
database = "../../resources/HR_Data/HR_Attrition.db"
conn_string = f"sqlite:///{database}"
```

```python
engine = create_engine(conn_string)
```

```python
# get all tables
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```

```python
min_age = 20
max_age = 30
gender = '"Male", "Female"'
attrition = '"No", "Yes"'

query = f"""
        SELECT
            *
        FROM
            EmpPersonalInfo pi
        LEFT JOIN EmpWorkInfo wi ON
            pi.EmployeeID = wi.EmployeeID
        WHERE
            Age >= {min_age}
            AND Age <= {max_age}
            AND Gender in ({gender})
            AND Attrition in ({attrition});
            """

print(query)

df = pd.read_sql(query, conn_string)
df.head()
```
